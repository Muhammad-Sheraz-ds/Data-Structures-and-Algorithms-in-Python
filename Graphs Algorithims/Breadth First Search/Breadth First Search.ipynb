```python
import networkx as nx
import matplotlib.pyplot as plt
from collections import deque

class Graph_Node:
    def __init__(self, label=None, color=None, predecessor=None, distance=0):
        self.label = label
        self.color = color
        self.predecessor = predecessor
        self.distance = distance

class Vertex:
    def __init__(self, label=None):
        self.label = label
        self.vertices = []

class Graph:
    def __init__(self):
        self.vertices = []

    def add_vertex(self, label=None):
        new_vertex = Vertex(label)
        self.vertices.append(new_vertex)
        return new_vertex

    def add_edge(self, vertex1, vertex2):
        vertex1.vertices.append(vertex2)
        vertex2.vertices.append(vertex1)

    def bfs(self, source_vertex, visualize=False):
        for vertex in self.vertices:
            vertex.color = 'White'
            vertex.distance = float('inf')
            vertex.predecessor = None

        source_vertex.color = 'Gray'
        source_vertex.distance = 0
        source_vertex.predecessor = None

        queue = deque([source_vertex])

        if visualize:
            self.visualize_graph()

        while queue:
            current_vertex = queue.popleft()
            for neighbor in current_vertex.vertices:
                if neighbor.color == 'White':
                    neighbor.color = 'Gray'
                    neighbor.distance = current_vertex.distance + 1
                    neighbor.predecessor = current_vertex
                    queue.append(neighbor)

                    if visualize:
                        self.visualize_graph()

            current_vertex.color = 'Black'

            if visualize:
                self.visualize_graph()

    def visualize_graph(self):
        G = nx.Graph()

        for vertex in self.vertices:
            G.add_node(vertex.label, color=vertex.color)

            for neighbor in vertex.vertices:
                G.add_edge(vertex.label, neighbor.label)

        pos = nx.spring_layout(G)
        colors = [data['color'] for node, data in G.nodes(data=True)]

        nx.draw(G, pos, with_labels=True, node_color=colors, cmap=plt.cm.Blues, edge_color='black')
        plt.show()

    def draw_graph(self):
        G = nx.Graph()

        for vertex in self.vertices:
            G.add_node(vertex.label)

            for neighbor in vertex.vertices:
                G.add_edge(vertex.label, neighbor.label)

        pos = nx.spring_layout(G)
        nx.draw(G, pos, with_labels=True, node_color='skyblue', edge_color='black', font_color='black')
        plt.show()



    
```

```python
graph = Graph()

# Add vertices with labels
v1 = graph.add_vertex(label='a')
v2 = graph.add_vertex(label='b')
v3 = graph.add_vertex(label='c')
v4 = graph.add_vertex(label='d')
v5 = graph.add_vertex(label='e')
v6 = graph.add_vertex(label='f')
v7 = graph.add_vertex(label='g')

# Add edges
graph.add_edge(v1, v2)
graph.add_edge(v1, v3)
graph.add_edge(v2, v4)
graph.add_edge(v2, v5)
graph.add_edge(v3, v6)
graph.add_edge(v4, v7)
graph.add_edge(v5, v7)
graph.add_edge(v6, v7)
graph.draw_graph()
```

```python
# Perform BFS from v1 with visualization
graph.bfs(v1, visualize=True)
```

```python
print("\nBFS results:")
for vertex in graph.vertices:
    print(f"Vertex: {vertex.label}, Color: {vertex.color}, Distance: {vertex.distance}, Predecessor: {vertex.predecessor.label if vertex.predecessor else None}")

```
