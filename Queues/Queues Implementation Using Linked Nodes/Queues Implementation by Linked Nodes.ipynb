```python
from abc import ABC, abstractmethod
class Queue(ABC):
    @abstractmethod
    def __init__(self):
        pass

    @abstractmethod
    def __str__(self):
        pass

    @abstractmethod
    def enqueue(self,item):
        pass

    @abstractmethod
    def dequeue(self):
        pass

    @abstractmethod
    def is_empty(self):
        pass

    def is_full(self):
        pass

    @abstractmethod
    def size(self):
        pass

```

```python
class Node():
    def __init__(self,data,link=None):
        self.data = data
        self.link = link

class Queues:
    def __init__(self):
        self.top = None
        self.rear = None

    def enqueue(self,val):
        if self.is_empty():
            self.top = Node(val)
            self.rear = Node(val)
            
        else:
            self.rear.link = Node(val)
            self.rear = self.rear.link
        
            

    def dequeue(self):
        if not self.is_empty():
            cn = self.top
            self.top = self.top.link
            return cn.data
            



    def is_empty(self):
        if self.top==None:
            return True
        return False

    def __str__(self):
        s=''
        if not self.is_empty():
            cn = self.node
            while cn.link!=None:
                s += str(cn.data) + ' '
                cn = cn.link
            return s + str(cn.data)
        else:
            return 'empty node'
        
    def peek(self):
        if self.top:
            return self.rear.data
```

```python
def main():
    queue = Queues()

    print("Is queue empty?", queue.is_empty())

    try:
        queue.dequeue()
    except Exception as e:
        print("Dequeue error:", e)

    for i in range(1, 6):
        queue.enqueue(i)


    try:
        queue.enqueue(6)
    except Exception as e:
        print("Enqueue error:", e)

    print("Front element:", queue.peek())

    print("Queue elements:")
    while not queue.is_empty():
        print(queue.dequeue(), end=" ")

    print("\nIs queue empty?", queue.is_empty())

if __name__ == "__main__":
    main()
```
