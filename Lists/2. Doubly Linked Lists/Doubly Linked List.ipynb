```python
class doublyLinkedList:
    class Node:
        def __init__(self, data=None):
            self.data = data
            self.next = None
            self.prev = None

    class DLLIterator:
        def __init__(self, node=None):
            self.current = node

        def __eq__(self, rhs):
            return self.current == rhs.current

        def __ne__(self, rhs):
            return self.current != rhs.current

        def getObject(self):
            return self.current.data

        def __iter__(self):
            return self
        
        def __next__(self):
            if self.current.next is not None and self.current.next.next is not None:
                cur_node = self.current
                self.current = self.current.next
                return self
            else:
                raise StopIteration


        def remove(self):
            if self.current is not None and self.current.prev is not None and self.current.next is not None:
                self.current.next.prev = self.current.prev
                self.current.prev.next = self.current.next
                tmp = self.current
                self.current = self.current.next  # Move to the next node
                del tmp

    class DLLRIterator:
        def __init__(self, node=None):
            self.current = node

        def __eq__(self, rhs):
            return self.current == rhs.current

        def __ne__(self, rhs):
            return self.current != rhs.current

        def getObject(self):
            return self.current.data

        def __iter__(self):
            return self
        
        def __next__(self):
            if self.current.prev is not None and self.current.prev.prev is not None:
                cur_node = self.current
                self.current = self.current.prev
                return self
            else:
                raise StopIteration


    def __init__(self):
        self.head = self.Node()
        self.tail = self.Node()
        self.head.next = self.tail
        self.tail.prev = self.head

    def __iter__(self):
        return self.DLLIterator(self.head)

    def __reversed__(self):
        return self.DLLRIterator(self.tail)

    def begining(self):
        return self.DLLIterator(self.head)

    def end(self):
        return self.DLLIterator(self.tail.prev)

    def rbegining(self):
        return self.DLLRIterator(self.tail)

    def rend(self):
        return self.DLLRIterator(self.head.next)

    def add_node(self, o):
        t = self.Node(o)
        t.next = self.tail
        t.prev = self.tail.prev
        self.tail.prev.next = t
        self.tail.prev = t
        
    def delete_at_beginning(self):
        if self.head.next != self.tail:
            first_node = self.head.next
            second_node = first_node.next

            self.head.next = second_node
            second_node.prev = self.head

            del first_node
        else:
            print("List is empty. Nothing to delete.")
            
            
    def delete_at_end(self):
        if self.head.next != self.tail:
            last_node = self.tail.prev
            second_last_node = last_node.prev

            second_last_node.next = self.tail
            self.tail.prev = second_last_node

            del last_node
        else:
            print("List is empty. Nothing to delete.")



    def remove(self, index):
        cur = None
        if type(index) is int:
            cur = self.head.next
            i = 0
            while i != index and cur != self.tail:
                i = i + 1
                cur = cur.next
            if cur == self.tail:
                raise Exception("Invalid index")
        elif type(index) is self.DLLIterator:
            cur = index.current
        cur.prev.next = cur.next
        cur.next.prev = cur.prev
        del cur
        

    def Display(self):
        cur = self.head.next
        while cur != self.tail:
            print(cur.data, end=" ")
            cur = cur.next
        print()

    def insert(self, val, new_val):
        node = self.Node(new_val)
        t = self.head.next
        while t is not None:
            if t.data == val:
                node.prev = t
                node.next = t.next
                t.next.prev = node
                t.next = node
                return

    def traverse(self):
        cur = self.head.next
        while cur != self.tail:
            print(cur.data, end=" ")
            cur = cur.next
        print()
        
        

    def insertatbegin(self, k):
        ptr = self.Node(k)
        if self.head.next is None:
            self.head.next = ptr
            ptr.prev = self.head
            ptr.next = self.tail
            self.tail.prev = ptr
        else:
            temp = ptr
            temp.next = self.head.next
            self.head.next.prev = temp
            temp.prev = self.head
            self.head.next = temp

    def insertatend(self, k):
        ptr = self.Node(k)
        if self.head.next is None:
            self.head.next = ptr
            ptr.prev = self.head
            ptr.next = self.tail
            self.tail.prev = ptr
        else:
            temp = ptr
            temp.prev = self.tail.prev
            self.tail.prev.next = temp
            temp.next = self.tail
            self.tail.prev = temp

    def insertatind(self, k, pos):
        if pos < 1:
            print("Please enter a valid position")
            return
        elif pos == 1:
            self.insertatbegin(k)
        else:
            src = self.head.next
            i = 1
            while i < pos and src != self.tail:
                src = src.next
                i += 1
            if src == self.tail:
                print("Please enter a valid position")
                return
            ptr = self.Node(k)
            ptr.prev = src.prev
            ptr.next = src
            src.prev.next = ptr
            src.prev = ptr

    
    def insert_after(self, value, new_value):
        new_node = self.Node(new_value)
        current_node = self.head.next
        while current_node != self.tail:
            if current_node.data == value:
                new_node.prev = current_node
                new_node.next = current_node.next
                current_node.next.prev = new_node
                current_node.next = new_node
                if current_node == self.tail.prev:
                    self.tail.prev = new_node  # Update tail if inserting at the end
                return
            current_node = current_node.next

    def insert_before(self, value, new_value):
        new_node = self.Node(new_value)
        current_node = self.head.next
        while current_node != self.tail:
            if current_node.data == value:
                new_node.prev = current_node.prev
                new_node.next = current_node
                current_node.prev.next = new_node
                current_node.prev = new_node
                if current_node == self.head.next:
                    self.head.next = new_node  # Update head if inserting at the beginning
                return
            current_node = current_node.next
  

        
            


            
def main():
    # Create a new Doubly Linked List
    dll = doublyLinkedList()

    # Insert elements into the list
    dll.add_node(10)
    dll.add_node(20)
    dll.add_node(30)
    dll.add_node(40)
    dll.add_node(50)

    # Display the initial list
    print("Initial Linked List:")
    dll.Display()

    # Insert an element at the beginning
    dll.insertatbegin(5)
    print("Linked List after inserting 5 at the beginning:")
    dll.Display()

    # Insert an element at the end
    dll.insertatend(60)
    print("Linked List after inserting 60 at the end:")
    dll.Display()

    # Insert an element at a specific position
    dll.insertatind(35, 4)
    print("Linked List after inserting 35 at position 4:")
    dll.Display()

    # Remove an element from a specific position
    dll.remove(2)  # Remove the element at position 2
    print("Linked List after removing the element at position 2:")
    dll.Display()

    # Iterate through the list using the custom iterator
    print("Iterating through the list:")
    for item in dll:
        print(item.getObject())

    # Reverse iteration using the custom reverse iterator
    print("Reverse Iterating through the list:")
    for item in reversed(dll):
        print(item.getObject())

    # Manually iterate and remove elements
    itr = iter(dll)
    print("Manually iterate and remove elements:")
    while True:
        try:
            item = next(itr)
            print("Current element:", item.getObject())
            item.remove()
            print("Removed the current element.")
        except StopIteration:
            break

    # Display the final list after removals
    print("Linked List after manual removals:")
    dll.Display()

    # Insert an element after a specific value
    dll.insert_after(50, 60)
    print("Linked List after inserting 60 after 50:")
    dll.Display()

    # Insert an element before a specific value
    dll.insert_before(10, 5)
    print("Linked List after inserting 5 before 10:")
    dll.Display()

    # Delete a node at the beginning
    dll.delete_at_beginning()
    print("Linked List after deleting the first node:")
    dll.Display()

    # Delete a node at the end
    dll.delete_at_end()
    print("Linked List after deleting the last node:")
    dll.Display()


if __name__ == "__main__":
    main()
   
```

# Doubly Linked List Functions - Time and Space Complexity

### `__init__(self)`

**Time Complexity:** O(1)
- The initialization method creates two sentinel nodes (head and tail) and sets them as the first and last elements in the list. This is a constant-time operation.

**Space Complexity:** O(1)
- Only two sentinel nodes are created regardless of the size of the list, so the space complexity is constant.

### `add_node(self, o)`

**Time Complexity:** O(1)
- Adding a node to the end of the list involves updating the prev and next pointers of the tail node and the new node. This is a constant-time operation.

**Space Complexity:** O(1)
- The space complexity is constant because only one additional node is created.

### `delete_at_beginning(self)`

**Time Complexity:** O(1)
- Removing the first node in the list involves updating the next pointer of the head node to point to the new first node. This is a constant-time operation.

**Space Complexity:** O(1)
- No additional data structures are created, so the space complexity is constant.

### `delete_at_end(self)`

**Time Complexity:** O(1)
- Removing the last node in the list involves updating the next pointer of the second-to-last node to point to the tail node. This is a constant-time operation.

**Space Complexity:** O(1)
- No additional data structures are created, so the space complexity is constant.

### `remove(self, index)`

**Time Complexity:** O(n)
- The time complexity of removing a node by its index depends on the position of the node to be removed. In the worst case, it may need to traverse the entire list, making it a linear time operation.

**Space Complexity:** O(1)
- No additional data structures are created, so the space complexity is constant.

### `insert(self, val, new_val)`

**Time Complexity:** O(n)
- Inserting a node after a specific value requires traversing the list to find the value to insert after. In the worst case, it may need to traverse the entire list, making it a linear time operation.

**Space Complexity:** O(1)
- No additional data structures are created, so the space complexity is constant.

### `insertatbegin(self, k)`

**Time Complexity:** O(1)
- Inserting a node at the beginning involves updating the next and prev pointers of the new node, the current first node, and the head node. This is a constant-time operation.

**Space Complexity:** O(1)
- The space complexity is constant because only one additional node is created.

### `insertatend(self, k)`

**Time Complexity:** O(1)
- Inserting a node at the end involves updating the next and prev pointers of the new node, the current last node, and the tail node. This is a constant-time operation.

**Space Complexity:** O(1)
- The space complexity is constant because only one additional node is created.

### `insertatind(self, k, pos)`

**Time Complexity:** O(n)
- Inserting a node at a specific position requires traversing the list to find the insertion point. In the worst case, it may need to traverse the entire list, making it a linear time operation.

**Space Complexity:** O(1)
- No additional data structures are created, so the space complexity is constant.

### `insert_after(self, value, new_value)`

**Time Complexity:** O(n)
- Inserting a node after a specific value requires traversing the list to find the value to insert after. In the worst case, it may need to traverse the entire list, making it a linear time operation.

**Space Complexity:** O(1)
- No additional data structures are created, so the space complexity is constant.

### `insert_before(self, value, new_value)`

**Time Complexity:** O(n)
- Inserting a node before a specific value requires traversing the list to find the value to insert before. In the worst case, it may need to traverse the entire list, making it a linear time operation.

**Space Complexity:** O(1)
- No additional data structures are created, so the space complexity is constant.

### `Display(self)`

**Time Complexity:** O(n)
- Displaying the entire list requires traversing and printing all elements in the list, making it a linear time operation.

**Space Complexity:** O(1)
- No additional data structures are created for displaying the list, so the space complexity is constant.

### `traverse(self)`

**Time Complexity:** O(n)
- Traversing the list to print its elements is a linear time operation, as it involves visiting all nodes.

**Space Complexity:** O(1)
- No additional data structures are created for traversal, so the space complexity is constant.

### `begining(self), end(self), rbegining(self), rend(self)`

**Time Complexity:** O(1)
- These methods return iterators pointing to the head, tail, or sentinel nodes. They are constant-time operations.

**Space Complexity:** O(1)
- These methods return existing iterator objects, so the space complexity is constant.

### Iterator Methods (`__iter__(self)`, `__reversed__(self)`, `getObject(self)`, `__next__(self)`, `remove(self)`

**Time Complexity:** O(1)
- These methods are used to create and manage iterators, so they are constant-time operations.

**Space Complexity:** O(1)
- These methods involve working with existing iterator objects, so the space complexity is constant.
