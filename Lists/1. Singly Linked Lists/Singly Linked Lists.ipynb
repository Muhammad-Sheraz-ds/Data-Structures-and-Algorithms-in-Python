```python
class MyLinkedList:
    class Node:
        def __init__(self, value=None, next_node=None):
            self.value = value
            self.next_node = next_node

    def __init__(self):
        self.head = None

    class LinkedListIterator:
        def __init__(self, node):
            self.current = node

        def __next__(self):
            if self.current is not None:
                current_node = self.current
                self.current = self.current.next_node
                return current_node.value
            else:
                raise StopIteration

    def __iter__(self):
        return self.LinkedListIterator(self.head)

    def append(self, value):
        new_node = self.Node(value)
        if self.head is None:
            self.head = new_node
        else:
            current_node = self.head
            while current_node.next_node is not None:
                current_node = current_node.next_node
            current_node.next_node = new_node

    def insert_at_beginning(self, value):
        new_node = self.Node(value)
        if self.head is None:
            self.head = new_node
        else:
            new_node.next_node = self.head
            self.head = new_node

    def insert_at_index(self, value, index):
        new_node = self.Node(value)
        current_node = self.head
        position = 0
        if position == index:
            self.insert_at_beginning(value)
        else:
            while current_node is not None and position + 1 != index:
                position = position + 1
                current_node = current_node.next_node

            if current_node is not None:
                new_node.next_node = current_node.next_node
                current_node.next_node = new_node
            else:
                print("Index not present")

    def insert_at_end(self, value):
        new_node = self.Node(value)
        if self.head is None:
            self.head = new_node
        else:
            current_node = self.head
            while current_node.next_node:
                current_node = current_node.next_node
            current_node.next_node = new_node

    def update_node(self, value, index):
        current_node = self.head
        position = 0
        while current_node is not None and position != index:
            position = position + 1
            current_node = current_node.next_node

        if current_node is not None:
            current_node.value = value
        else:
            print("Index not present")

    def remove_first_node(self):
        if self.head is not None:
            self.head = self.head.next_node

    def remove_last_node(self):
        if self.head is not None:
            current_node = self.head
            while current_node.next_node and current_node.next_node.next_node:
                current_node = current_node.next_node
            current_node.next_node = None

    def remove_at_index(self, index):
        if self.head is not None:
            current_node = self.head
            position = 0
            if position == index:
                self.remove_first_node()
            else:
                while current_node is not None and position + 1 != index:
                    position = position + 1
                    current_node = current_node.next_node

                if current_node is not None:
                    current_node.next_node = current_node.next_node.next_node
                else:
                    print("Index not present")

    def remove_node(self, value):
        current_node = self.head
        while current_node is not None and current_node.next_node and current_node.next_node.value != value:
            current_node = current_node.next_node

        if current_node is not None and current_node.next_node:
            current_node.next_node = current_node.next_node.next_node

    def size_of_linked_list(self):
        size = 0
        current_node = self.head
        while current_node:
            size = size + 1
            current_node = current_node.next_node
        return size

    def display(self):
        current_node = self.head
        while current_node is not None:
            print(current_node.value)
            current_node = current_node.next_node
        print()
        
    




def main():
    my_linked_list = MyLinkedList()
    
    # Appending values to the linked list
    my_linked_list.append(10)
    my_linked_list.append(20)
    my_linked_list.append(30)
    my_linked_list.append(70)
    my_linked_list.append(50)
    my_linked_list.append(60)
    my_linked_list.append(80)
    my_linked_list.append(90)
    my_linked_list.append(40)
    # Displaying linked list using Matplotlib
    my_linked_list.display()

if __name__ == "__main__":
    main()
```

# Time and Space Complexity Analysis for MyLinkedList Class

## `__init__(self)`
**Time Complexity:** O(1)

Initializing an empty linked list is a constant-time operation as it only involves setting the head reference to None.

**Space Complexity:** O(1)

The space complexity is constant because it only creates a reference to the head node.

## `append(self, value)`
**Time Complexity:** O(n)

Appending a value to the end of the linked list requires traversing the entire list to find the last node. In the worst case, it takes linear time.

**Space Complexity:** O(1)

Only one new node is created, so the space complexity is constant.

## `insert_at_beginning(self, value)`
**Time Complexity:** O(1)

Inserting a value at the beginning involves creating a new node and updating the head reference, which is a constant-time operation.

**Space Complexity:** O(1)

The space complexity is constant as it only creates one new node.

## `insert_at_index(self, value, index)`
**Time Complexity:** O(n)

Inserting a value at a specific index may require traversing the list until reaching the desired position. In the worst case, it takes linear time.

**Space Complexity:** O(1)

It creates only one new node, so the space complexity is constant.

## `insert_at_end(self, value)`
**Time Complexity:** O(n)

Inserting a value at the end of the linked list requires traversing the entire list to find the last node. In the worst case, it takes linear time.

**Space Complexity:** O(1)

Only one new node is created, so the space complexity is constant.

## `update_node(self, value, index)`
**Time Complexity:** O(n)

Updating a node at a specific index requires traversing the list until reaching the desired position. In the worst case, it takes linear time.

**Space Complexity:** O(1)

No additional data structures are created, so the space complexity is constant.

## `remove_first_node(self)`
**Time Complexity:** O(1)

Removing the first node involves updating the head reference, which is a constant-time operation.

**Space Complexity:** O(1)

No additional data structures are created, so the space complexity is constant.

## `remove_last_node(self)`

**Space Complexity:** O(1)

No additional data structures are created, so the space complexity is constant.

**Time Complexity:** O(n)

Removing a node at a specific index may require traversing the list until reaching the desired position. In the worst case, it takes linear time.
