# SinglyLinkedList Implementation


## Introduction
Provide a brief introduction to the purpose of this code and what the `MyLinkedList` class does.

## Class Structure
Explain the class structure, its components, and their functionalities.

### Node
- The `Node` class represents individual elements in the linked list.
- It contains attributes for `value` and `next_node`.

### MyLinkedList
- The main class, `MyLinkedList`, represents the linked list and includes methods for manipulation.

## Class Methods
Describe each of the class methods and their functionalities.

### `__init__`
- Constructor method for initializing the linked list.

### `append(value)`
- Appends a new node with the given `value` to the end of the linked list.

### `insert_at_beginning(value)`
- Inserts a new node with the given `value` at the beginning of the linked list.

### `insert_at_index(value, index)`
- Inserts a new node with the given `value` at the specified index in the linked list.

### `insert_at_end(value)`
- Inserts a new node with the given `value` at the end of the linked list.

### `update_node(value, index)`
- Updates the value of a node at the specified index.

### `remove_first_node()`
- Removes the first node from the linked list.

### `remove_last_node()`
- Removes the last node from the linked list.

### `remove_at_index(index)`
- Removes a node at the specified index.

### `remove_node(value)`
- Removes the node with the specified value.

### `size_of_linked_list()`
- Returns the size of the linked list.

### `display()`
- Displays the linked list.

## Visualization (Optional)
If you wish to include the visualization part:

### `visualize()`
- Visualizes the linked list using NetworkX and Matplotlib.

## Main Function
The `main()` function demonstrates the usage of the `MyLinkedList` class by creating a linked list and displaying its contents.


```python
class MyLinkedList:
    class Node:
        def __init__(self, value=None, next_node=None):
            self.value = value
            self.next_node = next_node

    def __init__(self):
        self.head = None

    class LinkedListIterator:
        def __init__(self, node):
            self.current = node

        def __next__(self):
            if self.current is not None:
                current_node = self.current
                self.current = self.current.next_node
                return current_node.value
            else:
                raise StopIteration

    def __iter__(self):
        return self.LinkedListIterator(self.head)

    def append(self, value):
        new_node = self.Node(value)
        if self.head is None:
            self.head = new_node
        else:
            current_node = self.head
            while current_node.next_node is not None:
                current_node = current_node.next_node
            current_node.next_node = new_node

    def insert_at_beginning(self, value):
        new_node = self.Node(value)
        if self.head is None:
            self.head = new_node
        else:
            new_node.next_node = self.head
            self.head = new_node

    def insert_at_index(self, value, index):
        new_node = self.Node(value)
        current_node = self.head
        position = 0
        if position == index:
            self.insert_at_beginning(value)
        else:
            while current_node is not None and position + 1 != index:
                position = position + 1
                current_node = current_node.next_node

            if current_node is not None:
                new_node.next_node = current_node.next_node
                current_node.next_node = new_node
            else:
                print("Index not present")

    def insert_at_end(self, value):
        new_node = self.Node(value)
        if self.head is None:
            self.head = new_node
        else:
            current_node = self.head
            while current_node.next_node:
                current_node = current_node.next_node
            current_node.next_node = new_node

    def update_node(self, value, index):
        current_node = self.head
        position = 0
        while current_node is not None and position != index:
            position = position + 1
            current_node = current_node.next_node

        if current_node is not None:
            current_node.value = value
        else:
            print("Index not present")

    def remove_first_node(self):
        if self.head is not None:
            self.head = self.head.next_node

    def remove_last_node(self):
        if self.head is not None:
            current_node = self.head
            while current_node.next_node and current_node.next_node.next_node:
                current_node = current_node.next_node
            current_node.next_node = None

    def remove_at_index(self, index):
        if self.head is not None:
            current_node = self.head
            position = 0
            if position == index:
                self.remove_first_node()
            else:
                while current_node is not None and position + 1 != index:
                    position = position + 1
                    current_node = current_node.next_node

                if current_node is not None:
                    current_node.next_node = current_node.next_node.next_node
                else:
                    print("Index not present")

    def remove_node(self, value):
        current_node = self.head
        while current_node is not None and current_node.next_node and current_node.next_node.value != value:
            current_node = current_node.next_node

        if current_node is not None and current_node.next_node:
            current_node.next_node = current_node.next_node.next_node

    def size_of_linked_list(self):
        size = 0
        current_node = self.head
        while current_node:
            size = size + 1
            current_node = current_node.next_node
        return size

    def display(self):
        current_node = self.head
        while current_node is not None:
            print(current_node.value)
            current_node = current_node.next_node
        print()
        
    
        
    '''def visualize(self):
        G = nx.DiGraph()
        current_node = self.head
        while current_node:
            G.add_node(current_node.value)
            if current_node.next_node:
                G.add_edge(current_node.value, current_node.next_node.value)
            current_node = current_node.next_node

        # Create a separate figure and display the graph
        plt.figure(figsize=(8, 6))
        pos = nx.spring_layout(G)  # You can adjust the layout as needed
        nx.draw(G, pos, with_labels=False, node_size=500, node_color="skyblue", font_color="black", font_size=10)
        plt.axis('off')
        plt.show()'''




def main():
    my_linked_list = MyLinkedList()
    
    # Appending values to the linked list
    my_linked_list.append(10)
    my_linked_list.append(20)
    my_linked_list.append(30)
    my_linked_list.append(70)
    my_linked_list.append(50)
    my_linked_list.append(60)
    my_linked_list.append(80)
    my_linked_list.append(90)
    my_linked_list.append(40)
    # Displaying linked list using Matplotlib
    my_linked_list.display()

if __name__ == "__main__":
    main()
```
