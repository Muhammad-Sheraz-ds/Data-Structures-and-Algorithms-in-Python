```python
from abc import ABC,abstractmethod
class Stack(ABC):
    @abstractmethod
    def __init__(self):
        pass

    @abstractmethod
    def push(self,val):
        pass

    @abstractmethod
    def pop(self):
        pass

    @abstractmethod
    def is_empty(self):
        pass

    @abstractmethod
    def peek(self):
        pass

    @abstractmethod
    def size(self):
        pass
    
    @abstractmethod
    def is_full():
        pass
    
    @abstractmethod
    def __str__(self):
        pass

```
