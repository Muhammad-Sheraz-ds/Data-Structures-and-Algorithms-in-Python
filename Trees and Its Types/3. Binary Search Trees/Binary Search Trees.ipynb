```python
class BinSearchTree:
    def __init__(self):
        self.root = None

    class Node:
        def __init__(self,data,left=None,right=None):
            self.data = data
            self.right=right
            self.left=left

    class Iterator:
        def __init__(self,tree):
            self.stack = [tree]

        def __next__(self):
            if len(self.stack)>0:
                t = self.stack.pop()
                if t.right!=None:
                    self.stack.append(t.right)
                if t.left!=None:
                    self.stack.append(t.left)
                return t.data
            else:
                raise StopIteration

    def __iter__(self):
        return self.Iterator(self.root)        

    def insert(self,val):
        if self.root==None:
            self.root=self.Node(val)
            return
        self.aux_insert(self.root,val)

    def aux_insert(self,node, key):
        if node==None:
            node =  self.Node(key)
        elif key < node.data:
            node.left= self.aux_insert(node.left, key)

        elif key >= node.data:
            node.right = self.aux_insert(node.right, key)
        return node

    def Delete(self,val):
        
        self.root = self.Aux_Delete(self.root, val)

    def Aux_Delete(self,root,key):
        if root is None:
            return root
        if key < root.data:
            root.left = self.Aux_Delete(root.left, key)
        elif key > root.data:
            root.right = self.Aux_Delete(root.right, key)
        else:
            if root.left is None:
                return root.right
            elif root.right is None:
                return root.left
            root.data = self.minValueNode(root.right).data
            root.right = self.Aux_Delete(root.right, root.data)
        return root



    def search(self,val):
        t = self.root
        while t!=None:
            if t.data == val:
                return t
            if val < t.data:
                t = t.left
            else:
                t = t.right
        return f'Node with Value {val} is not Found'

    def minValueNode(self,rt=None):
        current = rt
        if not rt:
            current = self.root
        while current.left is not None:
            current = current.left
        return current

    def maxValueNode(self,rt=None):
        current = rt
        if not rt:
            current = self.root
        while current.right is not None:
            current = current.right
        return current


    def height(self):
        return self.aux_height(self.root)

    def aux_height(self,rt):
        if rt!=None:
            return max(self.aux_height(rt.left),self.aux_height(rt.right))+1
        return 0

    def Display(self):
        self.aux_Display(self.root)
        print()

    def aux_Display(self,rt):
        if rt!=None:
            print(rt.data)
            self.aux_Display(rt.left)
            self.aux_Display(rt.right)

    def Inorder_Traversal(self):
        self.Aux_Inorder_Traversal(self.root)
        print()
    def Aux_Inorder_Traversal(self,rt):
        if rt!=None:
            self.Aux_Inorder_Traversal(rt.left)
            print(rt.data)
            self.Aux_Inorder_Traversal(rt.right)

    def Inorder_Traversal(self):
        self.Aux_Inorder_Traversal(self.root)
        print()
    def Aux_Inorder_Traversal(self,rt):
        if rt!=None:
            self.Aux_Inorder_Traversal(rt.left)
            print(rt.data)
            self.Aux_Inorder_Traversal(rt.right)

    def Pre_Order_Traversal(self):
        self.Aux_Pre_Order_Traversal(self.root)
        print()

    def Aux_Pre_Order_Traversal(self,rt):
        if rt!=None:
            print(rt.data)
            self.Aux_Pre_Order_Traversal(rt.left)
            self.Aux_Pre_Order_Traversal(rt.right)

    def Post_Order_Traversal(self):
        self.Aux_Post_Order_Traversal(self.root)
        print()

    def Aux_Post_Order_Traversal(self,rt):
        if rt!=None:
            self.Aux_Post_Order_Traversal(rt.left)
            self.Aux_Post_Order_Traversal(rt.right)
            print(rt.data)

def main():
    bst = BinSearchTree()
    bst.insert(30)
    bst.insert(10)
    bst.insert(40)
    bst.insert(5)
    bst.insert(4)
    bst.insert(20)
    bst.insert(8)
    bst.insert(6)
    bst.insert(56)
    bst.insert(76)
    bst.insert(3)
    bst.insert(98)
    bst.insert(32)
    bst.insert(21)
    bst.insert(45)
    bst.Display()
    bst.Pre_Order_Traversal()
    bst.Delete(6)
    print('DELETE',6)
    bst.Pre_Order_Traversal()
    bst.Delete(4)
    print('DELETE',4)
    bst.Pre_Order_Traversal()
    bst.Delete(40)
    print('DELETE',40)
    bst.Pre_Order_Traversal()
    



main()

    
```

```python
def main():
    bst = BinSearchTree()
    
    insert_vals = [30, 10, 40, 5, 4, 20, 8, 6, 56, 76, 3, 98, 32, 21, 45]
    for val in insert_vals:
        bst.insert(val)
    
    print("Initial Binary Search Tree:")
    bst.Display()
    
    print("In-Order Traversal:")
    bst.Inorder_Traversal()
    
    print("Pre-Order Traversal:")
    bst.Pre_Order_Traversal()
    
    print("Post-Order Traversal:")
    bst.Post_Order_Traversal()

    print("Height of the Tree:", bst.height())
    
    vals = [6, 99, 20]
    for val in vals:
        result = bst.search(val)
        print(f"Search for {val}: {result}")
    
    vals = [6, 40, 99]
    for val in vals:
        print(f"Deleting node with value {val}")
        bst.Delete(val)
        print("Updated Binary Search Tree:")
        bst.Display()

if __name__ == "__main__":
    main()
```
