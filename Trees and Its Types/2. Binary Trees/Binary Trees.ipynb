```python
# linked nodes based code for implementation of Binary TREE basics
class BinTree:
    class Node:
        def __init__(self, data=None):
            self.data = data
            self.left = None
            self.right = None

    def __init__(self):
        self.root = None

    class PreIterator:
        def __init__(self, node):
            self.stk = [node]

        def __next__(self):
            if len(self.stk) > 0:
                cur = self.stk[-1]
                del self.stk[-1]
                if cur.left != None:
                    self.stk.append(cur.left)
                if cur.right != None:
                    self.stk.append(cur.right)
                return cur.data
            else:
                raise StopIteration

    def __iter__(self):
        return self.PreIterator(self.root)

    def addroot(self, n):
        if self.root == None:
            self.root = self.Node(n)
        else:
            raise Exception("Root aleady exist")

    def addleftchild(self, n, p=None):
        if p == None:
            raise Exception("Parent key required")
        elif n != None and p != None:
            if self.root.data == p and self.root.left == None:
                self.root.left = self.Node(n)
                return
            self.addleftchildAux(self.root, n, p)

    def addleftchildAux(self, c, n, p):
        if c != None:
            if c.data == p and c.left == None:
                c.left = self.Node(n)
                return
            else:
                self.addleftchildAux(c.left, n, p)
                self.addleftchildAux(c.right, n, p)

    def addrightchild(self, n, p=None):
        if p == None:
            raise Exception("Parent key required")
        elif n != None and p != None:
            if self.root.data == p and self.root.right == None:
                self.root.right = self.Node(n)
                return
            self.addrightchildAux(self.root, n, p)

    def addrightchildAux(self, c, n, p):
        if c != None:
            if c.data == p and c.right == None:
                c.right = self.Node(n)
                return
            else:
                self.addrightchildAux(c.left, n, p)
                self.addrightchildAux(c.right, n, p)

    def Display(self):
        self.DisplayAux(self.root, 0)
        print()

    def DisplayAux(self, c, level):
        if c != None:
            print("    " * level, c.data, sep="")
            self.DisplayAux(c.left, level + 1)
            self.DisplayAux(c.right, level + 1)

    def level_order_travsrsal(self):
        if self.root == None:
            return []
        self.levels = self.aux_levels(self.root, 0)
        self.res = [[] for j in range(self.levels + 1)]
        self.Aux_level_order_travsrsal(self.root, 0)
        for level in range(len(self.res)):
            print(f'Nodes of Level {level}:', end=' ')
            for node in range(len(self.res[level])):
                print(self.res[level][node], end=' ')
            print()

    def Inorder_Traversal(self):
        self.Aux_Inorder_Traversal(self.root)
        print()

    def Aux_Inorder_Traversal(self, rt):
        if rt != None:
            self.Aux_Inorder_Traversal(rt.left)
            print(rt.data)
            self.Aux_Inorder_Traversal(rt.right)

    def Inorder_Traversal(self):
        self.Aux_Inorder_Traversal(self.root)
        print()

    def Aux_Inorder_Traversal(self, rt):
        if rt != None:
            self.Aux_Inorder_Traversal(rt.left)
            print(rt.data)
            self.Aux_Inorder_Traversal(rt.right)

    def Pre_Order_Traversal(self):
        self.Aux_Pre_Order_Traversal(self.root)
        print()

    def Aux_Pre_Order_Traversal(self, rt):
        if rt != None:
            print(rt.data)
            self.Aux_Pre_Order_Traversal(rt.left)
            self.Aux_Pre_Order_Traversal(rt.right)

    def Post_Order_Traversal(self):
        self.Aux_Post_Order_Traversal(self.root)
        print()

    def Aux_Post_Order_Traversal(self, rt):
        if rt != None:
            self.Aux_Post_Order_Traversal(rt.left)
            self.Aux_Post_Order_Traversal(rt.right)
            print(rt.data)

    def remove(self, val):
        if self.root and self.root.data == val:
            self.root = None
            return
        self.Aux_Remove(self.root, val)

    def Aux_Remove(self, rt, val):
        if rt:
            if rt.left and rt.left.data == val:
                rt.left = None
                return
            if rt.right and rt.right.data == val:
                rt.right = None
                return
            self.Aux_Remove(rt.left, val)
            self.Aux_Remove(rt.right, val)

    def Reverse_level_order_travsrsal(self):
        if self.root == None:
            return []
        self.levels = self.aux_levels(self.root, 0)
        self.res = [[] for j in range(self.levels + 1)]
        self.Aux_level_order_travsrsal(self.root, 0)
        for level in range(len(self.res) - 1, -1, -1):
            print(f'Nodes of Level {level}:', end=' ')
            for node in range(len(self.res[level])):
                print(self.res[level][node], end=' ')
            print()

    def aux_levels(self, root, levels):
        if root != None:
            if root.left == None and root.right == None:
                return levels
            s1 = self.aux_levels(root.left, levels + 1)
            s2 = self.aux_levels(root.right, levels + 1)
            return max(s1, s2)
        return levels

    def Aux_level_order_travsrsal(self, rt, level):
        if rt:
            self.res[level].append(rt.data)
            self.Aux_level_order_travsrsal(rt.left, level + 1)
            self.Aux_level_order_travsrsal(rt.right, level + 1)

    def search(self,val):
        self.node = None
        self.Aux_search(self.root,val)
        if self.node:
            self.DisplayAux(self.node,0)
            return 
        print(f'Node with val {val} Not Found')
        
    def Aux_search(self,rt,val):
        if rt:
            if rt.data==val:
                self.node = rt
                return 
            self.Aux_search(rt.left,val)
            self.Aux_search(rt.right,val)


def main():
    t = BinTree()

    # add the root element to the BinTree
    t.addroot(10)
    # add the element as a child node to an existing node of BinTree
    t.addleftchild(30, 10)
    t.addrightchild(70, 10)
    t.addleftchild(90, 30)
    t.addrightchild(60, 30)
    t.addleftchild(50, 90)
    t.addleftchild(40, 70)
    t.addrightchild(20, 70)

    print("Display BinTree data:")
    t.Display()
    print()

    
    print("OUTPUT of 'for d in t:'")
    for d in t:
        print(d)
    print()


main()


```

```python
def main():
    t = BinTree()

    t.addroot(10)
    t.addleftchild(30, 10)
    t.addrightchild(70, 10)
    t.addleftchild(90, 30)
    t.addrightchild(60, 30)
    t.addleftchild(50, 90)
    t.addleftchild(40, 70)
    t.addrightchild(20, 70)

    print("Display BinTree data:")
    t.Display()
    print()

    print("Pre-Order Traversal:")
    t.Pre_Order_Traversal()
    print()

    print("In-Order Traversal:")
    t.Inorder_Traversal()
    print()

    print("Post-Order Traversal:")
    t.Post_Order_Traversal()
    print()

    print("Level Order Traversal:")
    t.level_order_travsrsal()
    print()

    print("Reverse Level Order Traversal:")
    t.Reverse_level_order_travsrsal()
    print()

    val = 90
    print(f"Removing node with value {val}:")
    t.remove(val)
    t.Display()
    print()
    
    val = 10
    print(f"Searching for node with value {val}:")
    t.search(val)

    t.addleftchild(15, 60)
    t.addrightchild(25, 60)

    print("In-Order Traversal after adding nodes:")
    t.Inorder_Traversal()
    print()

    val = 15
    print(f"Removing leaf node with value {val}:")
    t.remove(val)
    t.Display()
    print()

    val = 70
    print(f"Removing node with (value {val}):")
    t.remove(val)
    t.Display()
    print()

    val = 40
    print(f"Searching for node with value {val}:")
    t.search(val)

main()
```
